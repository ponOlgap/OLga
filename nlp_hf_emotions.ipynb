```python
from transformers import pipeline
```

Создаем pipeline для классификации текста в Hugging Face

```python
classifier = pipeline("text-classification", 
                      model="cointegrated/rubert-tiny2-cedr-emotion-detection")
```

**Поэкспериментируем с моделями для определения определения тональности текта!**

```python
s = [""]*3

s[0] = "Класс - Это формальная модель или шаблон, задающий структуру объекта. Объект - это ещё одна сущьность, которая создаётся на основе класса (штампуется на основе класса). Между классом и объектом такие же отношения, как между чертежом и изделием на его основе. Например, есть чертеж детали, по которому ее изготавливают. Очевидно, что на основе чертежа можно изготавить мнжество детелей. Поэтом, важно корректно проработать сам чертёж. Ошибки чертежа в конечном итоге скажутся на ошибках в каждой детали."
s[1] = "Пытается ли Германия сейчас хотя бы осознать, в какой угол ее загнали англосаксы с подрывом Северных потоков? Да, конечно, часть общества все больше обеспокоена тем, что Германии не оставляют вообще никакой свободы маневра, особенно на Востоке, — только идти по все более узкому коридору нарастающего конфликта с Россией из-за Украины"
s[2] = "Одеяло Убежало, Улетела простыня, И подушка, Как лягушка, Ускакала от меня. Я за свечку, Свечка — в печку!" 
results = [0]*3
for i in range(3):
  results[i] = classifier(s[i])
  print(results[i], " ТЕКСТ", classifier(s[i]))
  
```

```python
results
```

Выводим все возможные классы для детского стихотворения

```python
classifier(s[2],
           return_all_scores=True)
```

Все три текта не токсичны, но третий содержит противоречия (что очевидно :) )

**Поробуем найти ответ на вопрос**

```python
qa_model = pipeline("question-answering", model="AlexKay/xlm-roberta-large-qa-multilingual-finedtuned-ru")
```

```python
vopros = "Что такое объект?"
```

```python
result = qa_model(question = vopros, context = s[0])
```

```python
print(result)
```
